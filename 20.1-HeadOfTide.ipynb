The purpose of this code is to determine which catchments have tidal influence as determined by the NJ Head of Tide (HOT) Survey. HOT points have catchment data added in ArcGIS. We then create a map of downstream catchments using the NHDPlus Value Added Attribute (VAA) dataset.

```python
import dbfread as dbf
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',15)

# These are the attributes for the HUC02 of 02, containing 0201*, 0202*, 0203*, etc. 
# All of NJ is contained in this file
nhd_file = 'H:/My Drive/HAND/NHDPlusAttributes/PlusFlowlineVAA.dbf'
nhd = pd.DataFrame(iter(dbf.read(nhd_file)))

# csv_name = nhd_file[:-4]+'.csv'
# nhd.to_csv(csv_name,index=False)
```

```python
tidal = nhd[['ComID','Tidal']]
tidal.to_csv('H:/My Drive/HAND/NHDPlusAttributes/Tidal.csv',index=False)
file = 'H:/My Drive/ArcGIS-NJ-HAND/Tide/020301.csv'
reclass = pd.read_csv(file)
reclass = reclass.merge(tidal, left_on='FROM_', right_on='ComID', how='left')
reclass.to_csv(file[:-4] + 'tide-vaa.csv', index=False)
```
