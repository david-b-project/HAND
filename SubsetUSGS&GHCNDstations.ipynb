The purpose of this code is to subset USGS and GHCND data by geographic locations. Large tables containing national or global data are opened and bounded by coordinates to create subsets for the NJ region.

```python
import pandas as pd
```

```python
flow = pd.read_csv('./USGS Flow/Gages.csv',)
```

```python
flow
```

```python
flow_nj = flow.loc[(flow.POINT_Y < 42.50)  & \
                   (flow.POINT_Y > 38.90)  & \
                   (flow.POINT_X < -73.49) & \
                   (flow.POINT_X > -76.39)]
flow_nj
```

```python
# flow_nj.to_csv('./USGS Flow/GagesNJ.csv')
```

```python
precip = pd.read_csv('./PrecipData/ghcnd-stations.csv')
precip
```

```python
precip_NJ = precip.loc[(precip.Lat < 42.50)  & \
                       (precip.Lat > 38.90)  & \
                       (precip.Lon < -73.49) & \
                       (precip.Lon > -76.39)]
precip_NJ
```

```python
precip_NJ.to_csv('./PrecipData/ghcnd-stations-NJ.csv')
```
