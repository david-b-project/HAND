```python
data = pd.read_csv('./USGS Flow/NJ Stream Gages.csv',infer_datetime_format=True)
data
```

```python
data.dtypes
```

```python
for site in data.site_no.unique():
    print(site)
```

```python
maxflow = pd.DataFrame(data.site_no.unique(), columns=['site'])
maxflow['flow'] = ''
maxflow
```

```python
for site in maxflow.site:
    subset = data.loc[data.site_no==site]
    maxflow.flow.loc[maxflow.site == site] = subset['194265_00060'].max()
```

```python
maxflow
```

```python
usgs = pd.read_csv('./USGS Flow/Gages.csv')
usgs
```

```python
usgs2=usgs[['SOURCE_FEA', 'FLComID']]
usgs2
```

```python
usgs2.dtypes
```

```python
maxflow2 = maxflow.merge(usgs2, left_on='site', right_on='SOURCE_FEA')
maxflow2
```

```python
nwm = pd.read_csv('./USGS Flow/NWM_maxflow.csv')
nwm
```

```python
maxflow2 = maxflow2.merge(nwm, left_on='FLComID', right_on='COMID')
maxflow2
```

```python
maxflow2.to_csv('./USGS Flow/Gage&NWM.csv')
```

```python
ht = pd.read_csv('./USGS Flow/GageHeight.csv')
ht
```

```python
ht = ht.sort_values(by='194264_00065',ascending=False)
ht
```

```python
ht2 = ht.drop_duplicates(subset='site_no')
ht2
```

```python
nj = pd.read_csv('./USGS Flow/GagesNJ_HUC.csv')
nj
```

```python
ht2 = ht2.merge(nj, left_on='site_no', right_on='SOURCE_FEA',how='left')
ht2
```

```python
ht2.to_csv('./USGS Flow/GagesNJ_Height.csv')
```
